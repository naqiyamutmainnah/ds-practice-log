```python
#persiapkan semua library pada sel ini
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
#buka berkas tugas dan masukkan dalam suatu variabel/dataframe
#tampilkan datanya
data = pd.read_csv("text_cleaning.csv")
data.head()
```

```python
#buat sebuah fungsi case folding yang memiliki fungsi di bawah
'''
1. lowercase
2. menghilangkan angka
3. menghilangkan tanda baca
4. menghilangkan tautan
5. menghilangkan whitespaces
'''
import re

def case_folding(text):

  text = text.lower()

  text = re.sub(r'\d+', '', text)

  text = re.sub(r'[^\w\s]', '', text)

  text = re.sub(r'http\S+', '', text)

  text = text.strip()

  return text

data['artikel'] = data['artikel'].apply(case_folding)

data.head()
```

```python
#aplikasikan fungsi pada kolom artikel dan taruh hasilnya pada kolom clean
#tampilkan
def clean_text(text):

  text = re.sub(r'[^\w\s]', '', text)
  text = re.sub(r'\s+', ' ', text)
  return text.strip()

data['clean'] = data['artikel'].apply(clean_text)
data[['artikel', 'clean']]
```

```python
#buat sebuah fungsi untuk menghilangkan stopwords
'''
menghilangkan stopword bahasa indonesia + kata bandung dan kota
'''
def remove_stopwords(text):
  stop_words = set(stopwords.words('indonesian'))
  stop_words.update(['bandung', 'kota'])
  words = word_tokenize(text)
  filtered_text = ' '.join([word for word in words if word not in stop_words])
  return filtered_text

data['clean'] = data['clean'].apply(remove_stopwords)
data[['artikel', 'clean']]

data.head()
```

```python
#buat sebuah fungsi untuk melakukan stemming
def stemming(text):
  stemmer = PorterStemmer()
  words = word_tokenize(text)
  stemmed_text = ' '.join([stemmer.stem(word) for word in words])
  return stemmed_text

data['clean'] = data['clean'].apply(stemming)
data[['artikel', 'clean']]
```
