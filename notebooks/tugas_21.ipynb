Gunakan dataset Advertising yang merupakan dataset umum untuk prediksi penjualan berdasarkan pengeluaran iklan pada TV, Radio, dan Newspaper. Kita akan membandingkan tiga algoritma regresi: Linear Regression, Decision Tree Regression, dan Random Forest Regression.




````python
# https://raw.githubusercontent.com/selva86/datasets/master/Advertising.csv
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge

from sklearn import metrics

filepath = 'https://raw.githubusercontent.com/selva86/datasets/master/Advertising.csv'
df = pd.read_csv(filepath)
df.head()
````

````python
df.info()
````

````python
display(df.columns)

#'latitude', 'housing_median_age', 'total_rooms', 'median_income'

data = df[['TV', 'radio', 'newspaper', 'sales']]
````

````python
x = data.iloc[:,:-1]
y = data.iloc[:,-1:]

display(y.head())
display(x.head())
````

````python
x_scaled = StandardScaler().fit_transform(x)
y_scaled = StandardScaler().fit_transform(y)
````

````python
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.2, random_state=42)

print(X_train.shape)
print(X_test.shape)
````

````python
model = Lasso(alpha=0.5)
model.fit(X_train, y_train)
````

````python
print(model.coef_)
````

````python
y_pred = model.predict(X_test)
````

````python
r2 = metrics.r2_score(y_test, y_pred)
print(r2)
````
