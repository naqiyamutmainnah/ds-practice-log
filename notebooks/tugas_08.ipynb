Kamu diberikan sebuah DataFrame yang merepresentasikan data penjualan dari beberapa toko cabang yang berbeda. 4.   Mengisi missing value pada kolom store_manager dengan string "Unknown".
5.   Menghapus duplikasi data dengan mempertimbangkan semua kolom kecuali timestamp.
6.   Mengurutkan DataFrame berdasarkan store_location (ascending), product_category (ascending), dan revenue (descending).
7.   Menambahkan kolom discount_applied yang berisi True jika discount lebih dari 0, dan False jika tidak ada diskon.
8.   Menghitung total revenue per store_location dan per store_manager secara manual dan mengembalikan hasilnya sebagai DataFrame baru.
9.   Mengembalikan DataFrame yang telah diubah beserta informasi missing value sebelum dan sesudah transformasi.

```python
data = pd.read_csv('/content/data_tugas_dataWrangling_3.csv')
display(data.isna())
display(data.isna().sum())
```

```python
# median = data['profit'].median()
# data['profit'].fillna(median, inplace = True)
data['profit'] = data.groupby('product_category')['profit'].transform(lambda x: x.fillna(x.median()))
```

```python
# mean = data['units_sold'].median()
# data['units_sold'].fillna(median, inplace = True)
data['units_sold'] = data.groupby('store_location')['units_sold'].transform(lambda x: x.fillna(x.mean()))
```

```python
data['store_manager'].fillna("Unknown", inplace=True)
```

```python
display(data.duplicated().sum())
```

```python
data = data.drop_duplicates(subset=data.columns.difference(['timestamp']))
```

```python
# store_location (ascending), product_category (ascending), dan revenue (descending)
display(data.sort_values(["store_location", "product_category"]), data.sort_values("revenue", ascending=False))
```

```python
data['discount_applied'] = data['discount'] > 0
```

```python
total_revenue_per_location = data.groupby('store_location')['revenue'].sum().reset_index(name='total_revenue_per_location')
total_revenue_per_manager = data.groupby('store_manager')['revenue'].sum().reset_index(name='total_revenue_per_manager')
display(total_revenue_per_location)
display(total_revenue_per_manager)
```

```python
missing_values_after = data.isnull().sum()
display(missing_values_after)
```
